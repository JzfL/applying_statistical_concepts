### Question 1: Simple Linear Regression 

Let's set up our workspace and use the `Boston` dataset in the `ISLP` library. Print `Boston` to learn more about the dataset.

```python
# Import standard libraries
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib.pyplot import subplots
import statsmodels.api as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import specific objects
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

from statsmodels.stats.outliers_influence \
     import variance_inflation_factor as VIF
from statsmodels.stats.anova import anova_lm

from ISLP import load_data
from ISLP.models import (ModelSpec as MS,
                         summarize,
                         poly)
```

```python
# Load the "Boston" dataset using the "load_data" function from the ISLP package
Boston = load_data('Boston')
Boston
```

Before we fit and review model outputs, we should visualize our data. Review the code and plot, shown below. Answer the following questions:

_(i)_ What are the `medv` and `dis` variables being plotted? (Hint: review this [link](https://islp.readthedocs.io/en/latest/datasets/Boston.html))<br>

*`medv` is the median value of owner-occupied homes in $1000s <br>*
*`dis` is the weighted mean of distances to 5 Boston employment centres*

_(ii)_ What concept ‘defines’ the plotted line?<br>

*This linear regression line uses a single predictor variable, in this case, the `dis` to predict a quantitative response `medv` by assuming the relationship between them is linear*

```python
# Extract the variables
medv = Boston['medv'].values.reshape(-1, 1)
dis = Boston['dis'].values.reshape(-1, 1)

# Plot data
plt.scatter(dis, medv, label='Data')
plt.xlabel('dis')
plt.ylabel('medv')

# Fit a linear regression model
lm = LinearRegression()
lm.fit(dis, medv)

# Plot the regression line
plt.plot(dis, lm.predict(dis), color='red', label='Regression Line')

# Add labels and legend
plt.xlabel('dis')
plt.ylabel('medv')
plt.legend()

# Show the plot
plt.show()
```

Consider the variables plotted above. In the context of the `Boston` dataset:

_(iii)_ What is the (implied) null hypothesis? What is the (implied) alternative hypothesis?

*The null hypothesis in this case is that there's ***no*** relationship between medium household income and household relative distance from employment centres <br>
while the alternative hypothesis is that there ***is some positive*** relationship between the two*
 

_(iv)_ Now, let’s fit a simple regression model, using the general syntax `sm.OLS()` and `model.fit()`. As above, use `medv` as the response variable Y, and `dis` as the predictor variable X. (Hint: use the syntax `sm.add_constant` with the appropriate argument).

```python
# Add your code here
X = pd.DataFrame({'intercept': np.ones(Boston.shape[0]),
                  'dis': Boston['dis']})

y = Boston['medv']
model = sm.OLS(y, X)
results = model.fit()
summarize(results)
```

*16.784179 to 19.995997 for $B_0$* <br>
*0.721509 to 1.461717 for $B_1$*

### Question 2: Multiple Linear Regression 

We'll continue to use the `Boston` dataset for this question.

_(i)_ Fit a multiple linear regression, with two predictor variables: $X_1$ is `dis`, and $X_2$ is `rm`. As before, keep `medv` as the response variable Y. (Hint: use the syntax `sm.add_constant` with the appropriate arguments).

```python
# Add your code here
# Couldn't understand why use sm.add_constant
Xwithrm = pd.DataFrame({'intercept': np.ones(Boston.shape[0]),
                  'dis': Boston['dis'],
                  'rm': Boston['rm']})
Xwithrm[:5]

y = Boston['medv']
modelwithrm = sm.OLS(y, Xwithrm) 
resultswithrm = modelwithrm.fit()
summarize(resultswithrm)
resultswithrm.summary()

```

_(ii)_ In the context of the `Boston` dataset, state the null and alternative hypotheses.

*null hypotheses is the medium value of homes has no relation with the property's distance from employment centres and no relation with the property's number of rooms*<br>
*while the alternative hypotheses is there is some relationship between the values of home and its distance to employment centres and its number of rooms*

_(iii)_ Review the model output, using `summary()`. Does it appear that both `dis` and `rm` are predictive of `medv`? How did you determine this?

*Yes, both $p$ values are close to zero*

_(iv)_ We can use the inbuilt `sm.graphics.plot_regress_exog` function to generate helpful diagnostic plots (Hint: provide `plot_regress_exog` with the multiple regression model). Review the first generated plot, 'Residuals vs. Fitted'. Which observations are outliers? What impact might outliers have on our model?

*The observations on the top of 'Residuals VS Fitted' are outliers, these outliers will skew $Brm$ and $Bdis$ towards more positive direction*

```python
# Add your code here
modelq2 = sm.OLS.from_formula('medv ~ rm', data=Boston)
resultq2 = modelq2.fit()

fig, ax = plt.subplots(1, 2, figsize=(12, 10))

sm.graphics.plot_regress_exog(resultq2, 'rm', fig=fig)

ax[1].scatter(resultq2.fittedvalues, resultq2.get_influence().resid_studentized_internal, alpha=0.8)
ax[1].set_xlabel('Fitted Values')
ax[1].set_ylabel('Studentized Residuals')

plt.show()
```

```python
# Add your code here
modelq2dis = sm.OLS.from_formula('medv ~ dis', data=Boston)
resultq2dis = modelq2dis.fit()


fig, ax = plt.subplots(1, 2, figsize=(12, 10))


sm.graphics.plot_regress_exog(resultq2dis, 'dis', fig=fig)


ax[1].scatter(resultq2dis.fittedvalues, resultq2dis.get_influence().resid_studentized_internal, alpha=0.8)
ax[1].set_xlabel('Fitted Values')
ax[1].set_ylabel('Studentized Residuals')

plt.show()
```
